---
title: Plano en $\mathbb{R}^2$
---

Los puntos en el plano cartesiano los puntos se pueden representar como pares coordenados $(x,y)$, donde $x$ es la coordenada horizontal y $y$ es la coordenada vertical. En álgebra lineal,  en el plano podemos definir dos elementos importantes:

+ Vectores: Un vector en el plano es un objeto matemático que tiene una magnitud y una dirección. Se puede representar como un par ordenado de números reales $(x,y)$, donde $x$ es la componente horizontal y $y$ es la componente vertical del vector.

+ Puntos: Un punto en el plano es un objeto matemático que se representa como un par ordenado de números reales $(x,y)$, donde $x$ es la coordenada horizontal y $y$ es la coordenada vertical del punto.
